# Support Vector Regressor

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn import datasets
import pandas as pd
```

```python
data = datasets.load_diabetes(as_frame=True).frame
```

```python
data.head()
# data.shape
```

```python
X= data.drop(columns=["target"])
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size= 0.3, random_state = 42
)
```

```python
# Using this data trainsform into "2D" Dimensions
y_train.values.reshape(-1,1)
```

```python
y_scaler = StandardScaler()
# Using ((y_train.values.reshape(-1,1)) we convert Single column data into 2D then using (.ravel()) again Convert it into 1D
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1,1)).ravel()  # ravel used for flatted ( data convert into less dimensions) 
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1,1)).ravel()
```

```python
model = SVR()
model.fit(X_train,y_train_scaled)
```

```python
y_test_pred_scaled  = model.predict(X_test)
y_train_pred_scaled  = model.predict(X_train)
```

```python
print("train r2 :-",r2_score(y_train_scaled, y_train_pred_scaled))
print("test r2 :-",r2_score(y_test_scaled, y_test_pred_scaled))
```

### Linear Kernel

```python
model = SVR( kernel= "linear")
model.fit(X_train,y_train_scaled)
y_test_pred_scaled  = model.predict(X_test)
y_train_pred_scaled  = model.predict(X_train)
print("train r2 :-",r2_score(y_train_scaled, y_train_pred_scaled))
print("test r2 :-",r2_score(y_test_scaled, y_test_pred_scaled))
```

### Sigmoid Kernel (not Use ful for this condition)

```python
model = SVR( kernel= "sigmoid")
model.fit(X_train,y_train_scaled)
y_test_pred_scaled  = model.predict(X_test)
y_train_pred_scaled  = model.predict(X_train)
print("train r2 :-",r2_score(y_train_scaled, y_train_pred_scaled))
print("test r2 :-",r2_score(y_test_scaled, y_test_pred_scaled))
```

### Polynomial Kernel

```python
model = SVR( kernel= "poly")
model.fit(X_train,y_train_scaled)
y_test_pred_scaled  = model.predict(X_test)
y_train_pred_scaled  = model.predict(X_train)
print("train r2 :-",r2_score(y_train_scaled, y_train_pred_scaled))
print("test r2 :-",r2_score(y_test_scaled, y_test_pred_scaled))
```

# Hyperparameter Tunning Using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "C" : [1,2,3,5,10,30,20,50],
    "kernel" : ["rbf", "linear"],
    "epsilon" : [0.01, 0.1, 0.2, 0.3, 0.5]
}
```

```python
svr = SVR()
grid_search = GridSearchCV(svr, param_grid, scoring="r2", cv = 5)
grid_search.fit(X_train,y_train_scaled)
```

```python
print("Best Params :- ", grid_search.best_params_)
```

```python
best_model = SVR(C= 20, epsilon= 0.1, kernel= 'linear')
best_model.fit(X_train,y_train_scaled)
y_test_pred_scaled  = best_model.predict(X_test)
y_train_pred_scaled  = best_model.predict(X_train)
print("train r2 :-",r2_score(y_train_scaled, y_train_pred_scaled))
print("test r2 :-",r2_score(y_test_scaled, y_test_pred_scaled))
```
