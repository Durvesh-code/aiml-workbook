# Support Vector Classifier

```python
from sklearn.svm import SVC
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
data = datasets.load_iris(as_frame=True).frame
print(data.shape)
data.head()
```

```python
X = data.drop(columns=["target"])
y = data["target"]
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.4, random_state = 42, stratify = y
)
```

```python
scaler = StandardScaler()
X_train_Scaled = scaler.fit_transform(X_train)
X_test_Scalaed = scaler.transform(X_test)
```

# Build model with default RBF Kernel

```python
svc = SVC()
svc.fit(X_train_Scaled, y_train)
y_pred = svc.predict(X_test_Scalaed)
```

# Evaluate the model

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
print("Accuracy Score", accuracy_score(y_pred, y_test))
print("Classification Report", classification_report(y_test, y_pred))
```

# Build Model with Linear Kernal

```python
svc_linear = SVC(kernel="linear")
svc_linear.fit(X_train_Scaled, y_train)
y_pred = svc_linear.predict(X_test_Scalaed)
print("Accuracy Score", accuracy_score(y_test, y_pred))
```

# Build Model Polynomial Kernal

```python
svc_poly = SVC(kernel="poly")
svc_poly.fit(X_train_Scaled, y_train)
y_pred = svc_poly.predict(X_test_Scalaed)
print("Accuracy Score", accuracy_score(y_test, y_pred))
```

# Build Model sigmoid Kernal

```python
svc_sigmoid = SVC(kernel="sigmoid")
svc_sigmoid.fit(X_train_Scaled, y_train)
y_pred = svc_sigmoid.predict(X_test_Scalaed)
print("Accuracy Score", accuracy_score(y_test, y_pred))
```

```python
C_vals = [0.5, 1, 2, 3, 4, 5]
for c_vals in C_vals:
    svc = SVC(C = c_vals, kernel="rbf")
    svc.fit(X_train_Scaled, y_train)
    y_pred = svc.predict(X_test_Scalaed)
    print("C_vals = ", c_vals ,"Accuracy Score", accuracy_score(y_test, y_pred))
```
