```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(
    data = tips,
    x = 'total_bill',
    y = 'tip',
    hue = "smoker",
    # col= "day",
    size = "size",
    style = "smoker"
)
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(
    data = tips,
    x = 'total_bill',
    y = 'tip',
    kind= "line",
    hue = "smoker",
    # col= "day",
    size = "size",
    style = "smoker"
)
```

```python
x_vals = [i for i in range(1,11)]
y_vals = [i**2 for i in x_vals]
```

```python
sns.relplot(
    x = x_vals,
    y = y_vals,
    kind = "line"
)
```

# scatter plot

```python
sns.set_theme(None)

sns.scatterplot(
    data =tips,
    x = 'total_bill',
    y = 'tip',
    hue = 'time'
)
```

# Line plot

```python
sns.set_theme(style= "white")

sns.lineplot(
    data =tips,
    x = 'total_bill',
    y = 'tip',
    hue = 'time'
)
```

# Bar Plot

```python
sns.barplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex"
)
```

# Box Plot 

```python
sns.boxplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex"
)
```

# Histogram 

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.histplot(
    data = penguins,
    x = "body_mass_g",
    bins = 30
)
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights.count()
```

```python
flights_pivot = flights.pivot(index = "month", columns = "year",values = 'passengers')
flights_pivot
```

```python
sns.heatmap(
    flights_pivot,
    cmap = "coolwarm",
    annot = True,
    fmt = 'o'
)
plt.title("Passengers Heatmap")
plt.tight_layout()
```

# Best Practices

```python
fig, ax = plt.subplots()
sns.lineplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = 'sex',
    ax = ax,
    errorbar= None  # confidence interval will be hide
)
ax.set_title("Avg bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill Amout")
fig.tight_layout()
```
