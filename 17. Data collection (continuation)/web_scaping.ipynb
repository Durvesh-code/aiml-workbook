# BeautifulSoap 
Data extraction from Html

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
with open("Scraped_data/data1.html","r", encoding="utf-8") as f:
    html_content = f.read()
soup = BeautifulSoup(html_content,'lxml')
```

```python
soup
```

```python
all_h3 = soup.findAll("h3")
all_countries = []
for h3 in all_h3:
    name = h3.get_text(strip = True)
    # print(h3.find_parent('div')["class"])
    population = h3.find_next('div').select("span.country-population")[0].get_text(strip = True)
    # print(h3.find_next('div').select("span.country-population")[0].get_text(strip = True))
    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
```

```python
poulation_Data = pd.DataFrame(all_countries,columns=['Name','Population'])
```

```python
poulation_Data.to_csv("Scraped_data/poulation_Data.csv", index=False)
```
