# Bar charts

#### A bar chart is graph that displays categorical data using recatangular bars.
#### each bar's (height or length, in horizontal bars) represents a value.

```python
movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]
plt.bar(years,revenue,color = 'green')
plt.title("movies revenue in each year")
plt.xlabel("movies")
plt.ylabel("Revenue($M)")
for i in range(len(years)):
    plt.text(years[i],revenue[i]+10,str(revenue[i]),ha="center")
```

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]
years = [2008, 2009, 2010, 2011, 2012]

## core logic
x = np.arange(len(years))
width = 0.4

plt.bar(x - width/2,oscar_revenue,width,label="oscal movies")
plt.bar(x + width/2,non_oscar_revenue,width,label = "non oscar movies")
plt.title("Oscar Vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")
plt.legend()
```

# Scatter plot

#### A scatter plot shows the relationship between two variables by plotting points on a 2D plane.

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x < 135 else "red" for x in bp]

plt.scatter(age,bp,cmap="plasma",c=bp,s=bp)
plt.title("Age Vs Bp")
plt.xlabel("Age")
plt.ylabel("Bp")
plt.grid()
plt.colorbar(label="blood_pressure")
for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext =(age[i]+1,bp[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+5)
```

```python
# Multiple Plots
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (°C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (°C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]
```

```python
plt.scatter(winter_temp,winter_humidity,label = "Winter")
plt.scatter(summer_temp,summer_humidity,label = "Summer")
plt.title("Summer Vs Winter Data")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
for i in range(len(cities)):
    plt.annotate(cities[i],xy=(winter_temp[i],winter_humidity[i]),xytext =(winter_temp[i]+1,winter_humidity[i]+1))
for i in range(len(cities)):
    plt.annotate(cities[i],xy=(summer_temp[i],summer_humidity[i]),xytext =(summer_temp[i]+1,summer_humidity[i]+1))
plt.legend()
plt.colorbar()
```
