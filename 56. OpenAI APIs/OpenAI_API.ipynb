# AI Personal Assistant

```python
# query
# email summarization

class PersonalAssistant:
    def __init__(self):
        print("Hi, i am your personal assistant JatUUUUUUUUU")

    def ans_query(self):
        question = input("Ask the question")
        response = client.responses.create(
            model= "gpt-5.4",
            input= [
                {"role":"system", "content" : "Act like a helpful personal Assistant"},
                {"role" : "user", "content":question}
            ],
            temperature= 0.7,
            max_output_tokens= 512
        )
        print(response.output_text.strip())
    def summarize_email(self):
        question = input("Paste your gmail :-")
        prompt = f"summarize the following email into 2 to 3 sentences: {question}"
        response = client.responses.create(
            model= "gpt-5.4",
            input= [
                {"role":"system", "content" : "Act like an expert email assistant"},
                {"role" : "user", "content":question}
            ],
            temperature= 0.1,
            max_output_tokens= 512
        )
        print(response.output_text.strip())
        
```

```python
assistant = PersonalAssistant()
```

```python
assistant.ans_query()
```

```python
assistant.summarize_email()
```

# Generate Images

### Using Responses Api

```python
import base64
# images => base64 encoded string => decode into redeable strings
client = OpenAI( api_key = OPENAI_API_KEY)
```

```python
response = client.responses.create(
    model="gpt-5",
    input="Generate image of a highly detailed futurustic space ship, flying near a nebula",
    tools = [{"type":"image_generation"}]
)
image_data = [
    output.result
    for output in response.output
    if output.type == "image_generation_call"
]

# decode & store image in  a file
if image_data:
    image_base64 = image_data[0]
    with open("SpaceShip.png","wb") as f:   # wb ===> write binary
        image_bytes = base64.b64decode(image_base64)
        f.write(image_bytes)
```

```python
prompt = "Generate image of a space ship, flying near a nebula"

response = client.images.generate(
    model="gpt-image-2",
    prompt=prompt
)

image_base64 = response.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("SpaceShip.png","wb") as f:   # wb ===> write binary
    f.write(image_bytes)
```

## Edit the image

```python
prompt = "take reference of image and convert space ship into red color"

response = client.images.generate(
    model="gpt-image-2",
    prompt=prompt,
    image = [
        open("SpaceShip.png","rb")   # read binary
    ]
)

image_base64 = response.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("SpaceShip_red.png","wb") as f:   # wb ===> write binary
    f.write(image_bytes)
```
