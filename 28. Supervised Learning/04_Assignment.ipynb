# Assignment

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = sns.load_dataset("iris")
data = data.drop_duplicates()
data
```

```python
data["species"].unique()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
```

#  KNN Prediction

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = data.drop(columns=["species"])
y = data["species"]
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.5, random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=7)
```

```python
classifier.fit(X_train_scaled,y_train)
```

```python
y_pred = classifier.predict(X_test_scaled)
y_pred
print("Accuracy :-", accuracy_score(y_test,y_pred))
print("Precision (macro):", precision_score(y_test, y_pred, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
```

# pipeline and cross validation

```python
from sklearn.pipeline import Pipeline
classifier = KNeighborsClassifier()
pipe = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier())
    ]
)
param_grid = {'knn__n_neighbors':[3,5,7,9]}
classifierCV = GridSearchCV(
    pipe,
    param_grid,
    cv = 5,
    scoring = "%recall"
)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Accuracy :-", accuracy_score(y_test,y_pred))
print("Precision (macro):", precision_score(y_test, y_pred, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
```

# Logistic ReGression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.5, random_state=42
)
```

```python
pipe = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('logreg', LogisticRegression(max_iter=1000,solver='lbfgs'))
    ]
)
param_grid = {'logreg__C': np.logspace(-4,1,50)}
classifierCV = GridSearchCV(
    pipe,
    param_grid,
    cv = 5,
    scoring = "recall_macro",
    n_jobs=-1
)
classifierCV.fit(X_train,y_train)
y_pred = classifierCV.predict(X_test)
print("Accuracy :-", accuracy_score(y_test,y_pred))
print("Precision (macro):", precision_score(y_test, y_pred, average='macro'))
print("Recall (macro):", recall_score(y_test, y_pred, average='macro'))
print("Best CV score (recall_macro):", classifierCV.best_score_)
print("Best CV score (recall_macro):", classifierCV.best_params_)
```
