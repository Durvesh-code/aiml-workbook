# IsoLation Forest Code :-

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
# from sklearn.
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("thyroid_dataset.csv")
data.head()
```

```python
X = data.drop(columns= "Outlier_label")
y = data["Outlier_label"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(
    n_estimators = 200,
    contamination = 'auto',
    random_state = 42
)
```

```python
labels = clf.fit_predict(X_scaled)
labels
```

```python
# Visualize
from sklearn.decomposition import PCA

pca = PCA(
    n_components = 2
)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize = (8,6))

plt.scatter(x = X_pca[:,0], y = X_pca[:,1], c = labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("thyroid dataset")
```

```python
import numpy as np
n_outliers = np.sum(labels == -1)
n_normals = np.sum(labels == 1)
print("Outliers = ", n_outliers)
print("Normal = ", n_normals)
```

# LOF (Local Outlier Factor) Code :-

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
neighbors = LocalOutlierFactor(
    contamination = 0.036
)
labels = neighbors.fit_predict(X_scaled)
```

```python
plt.figure(figsize = (8,6))

plt.scatter(x = X_pca[:,0], y = X_pca[:,1], c = labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("thyroid dataset")
```

```python
import numpy as np
n_outliers = np.sum(labels == -1)
n_normals = np.sum(labels == 1)
print("Outliers = ", n_outliers)
print("Normal = ", n_normals)
```
