# Activity
From Quotes, extract all quotes & authors from all pages with the life tag

### Scrape Multiple page

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# url = f"https://quotes.toscrape.com/page/1/"
page_count = 1
while True:
    url = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(url)
    soup = BeautifulSoup(res.text,"lxml")
    quotes = soup.select("div.quote")
    if not quotes:
        print("no Valid pages anymore..")
        break
    with open(f"Scraped_data/quotes{page_count}.html","w",encoding = "utf-8") as f:
        f.write(res.text)
        print(f"dw data from page {page_count}")
    page_count += 1
```

### get use ful info from scrapped data

```python
with open("Scraped_data/quotes1.html","r",encoding = "utf-8") as f:
    html_content = f.read()
soup = BeautifulSoup(html_content,"lxml")
all_quotes = soup.select("div.quote")
print(all_quotes)
```

```python
life_quotes = []
for q in all_quotes:
    all_tags = []
    # print(all_tags)
    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())
        # print(tag.get_text(strip=True))

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text,author])
    print(life_quotes)
```

```python
pd.DataFrame(life_quotes,columns=(["Quotes","Author"]))
```

# Ai Response Better Code (Activity 2)

```python

import pandas as pd
from bs4 import BeautifulSoup

page_count = 1
life_quotes1 = []

while True:
    try:
        with open(f"Scraped_data/quotes{page_count}.html", "r", encoding="utf-8") as f:
            html_content = f.read()
    except FileNotFoundError:
        break  # stop if file doesn't exist
    
    soup = BeautifulSoup(html_content, "lxml")
    all_quotes1 = soup.select("div.quote")
    
    if not all_quotes1:
        break  # stop if no quotes found
    
    for q in all_quotes1:
        tags = [tag.get_text() for tag in q.select(".tags .tag")]
        if "life" in tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes1.append([text, author])
    
    page_count += 1

df = pd.DataFrame(life_quotes1, columns=["Quotes", "Author"])
df
```
