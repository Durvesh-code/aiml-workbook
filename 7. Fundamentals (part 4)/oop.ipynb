# Practice Problem (Instance, Class, Static Method)

```python
class product:
    count = 0
    def __init__(self,name,price):
        self.name = name
        self.price = price
        product.count +=1
    def get_info(self): # instance method
        print(f"Name of Product : {self.name}\nPrice of Product {self.price}\n")
    
    @classmethod
    def get_count(self):
        print(f"Total product in Store {self.count}")

    @staticmethod
    def get_discount(price,discount):
        print(f"Dicounted Price = {price - (price * discount/100)}")

p1 = product("phone",10000)
p2 = product("laptop",50000)

p1.get_info()
p2.get_info()
product.get_count()

# calc Discount
p1.get_discount(p1.price,10)
```

# 1. EncapSulation

```python
# Wrapping of Data and Function into Single Unit
""" 

"""
class BankAccount:
    def __init__(self,name,balance,id):
        self.name = name # Public Attribute
        self._balance = balance # Protected Attribute
        self.__id = id  # Private Attribute (Data Mangling) 

    def get_id(self):   # getter
        return self.__id
    
    def set_id(self,NewID):
        self.__id = NewID 

acc1 = BankAccount("Rahul",100_000,1)

acc1.set_id(200_000)
print("protected",acc1.name,acc1._balance)
print("private",acc1.name, acc1.get_id())
```

# Multiple Inheritance

```python
class Teacher:
    def __init__(self,salary):
        self.salary = salary
    
class Student:
    def __init__(self,gpa):
        self.gpa = gpa

class TA(Teacher,Student):
    def __init__(self, salary,gpa,name):
        super().__init__(salary)
        Student.__init__(self,gpa)
        self.name = name

ta1 = TA(15_000,9.3,"Durvesh")
print(ta1.name,ta1.gpa,ta1.salary)
```
