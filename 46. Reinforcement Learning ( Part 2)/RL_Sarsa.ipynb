# Sarsa

```python
gamma = 0.99
alpha = 0.5
epsilon = 0.1  # 10% explore and 90% exploitation
episodes = 1000
```

```python
# Q- table => store Q-values # it is grid of 48,4 dimensions  In Q table we store expected future returns 
Q = np.zeros((48,4))
```

```python
# Policy - episilon-greedy : state -> action
def epsilon_greedy(state):
    if random.random() < epsilon:
        return env.action_space.sample()  # random Action => Explore
    else:
        return np.argmax(Q[state])  # exploit (uses learning steps)
```

```python
for episode in range(episodes):
    render= (episode % 250 ==0)
    if render:
        env = gym.make("CliffWalking-v1", render_mode = "human")
        print(f"rendering env for episodes = {episode+1}/500")
    else:
        env = gym.make("CliffWalking-v1")
        
    done = False # When the episode is ended
    state, _ = env.reset()
    action = epsilon_greedy(state)
    episode_len = 0
    total_reward = 0
    while not done:
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        next_action = epsilon_greedy(next_state)

        # sarsa update
        Q[state, action] += alpha * (reward + gamma * (Q[next_state, next_action]- Q[state, action]))

        state = next_state
        action = next_action

        total_reward += reward
        episode_len += 1

    print(f"episode= {episode +1}/500 tot reward = {total_reward} & ep length = {episode_len} ")
    env.close()
```

# what did our model learn?

```python
env = gym.make("CliffWalking-v1",render_mode="human")
state, _ = env.reset()
done = False
total_reward = 0
episode_len = 0 
while not done:
    action = np.argmax(Q[state])
    state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated

    total_reward += reward
    episode_len += 1

print(f"total reward ={total_reward} & episode len ={episode_len} ")
env.close()
```
