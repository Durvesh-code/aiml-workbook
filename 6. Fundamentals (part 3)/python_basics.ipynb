# Lists

```python
# lists are mutable 
marks = [99,89,100,43,33]
print(marks[0])
print(len(marks)-1)
print(type(marks))
```

```python
# Slice 
print(marks[2:3])
```

```python
marks.append(4)
marks
```

```python
marks.insert(2,10)
marks
```

```python
marks.sort()
marks
```

```python
marks.reverse()
marks
```

# Tuple

```python
# Immutable in Nature    stored in format () 
tup = (1,2,3,4,5,2,6)
print(tup)
```

```python
tup.index(2)
```

```python
tup.count(2)   # how many times 2 repeated in the tuple
```

# palindrome string

```python
var1 = 'madam'
var_rev = var1[::-1]
if var1 == var_rev:
    print("palindrome string")
else:
    print("not a palindrome string")
```

```python
list1 = [1, 2, 7]
list2 = [2, 4, 5]
print(list1 + list2)
```
